```python
# EE.IMAGE
imagem1 = ee.Image(imgCol.first())
imagem2 = imgCol.get(2)

print(imagem1)
print(imagem1.getInfo())
print(imagem1.getInfo()['properties'])
print(imagem1.getInfo()['properties']['IMAGE_QUALITY']) #Sincrono
print(imagem1.get('IMAGE_QUALITY').getInfo()) #Assincrono 
```

```python
#Visualizando dados em um mapa
import folium # Biblioteca para visualização de dados

# Este método adiciona uma camada(ee.Image) ao mapa
def add_ee_layer(self, ee_image_object, vis_params, name):
  map_id_dict = ee.Image(ee_image_object).getMapId(vis_params)
  folium.raster_layers.TileLayer(
    tiles = map_id_dict['tile_fetcher'].url_format,
    attr = 'Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    name = name,
    overlay = True,
    control = True
  ).add_to(self)

# Este método adiciona uma camada ao mapa
def add_ee_layer_vect(self, ee_image_object, vis_params, name):
  map_id_dict = ee.FeatureCollection([ee_image_object]).getMapId(vis_params)
  folium.raster_layers.TileLayer(
    tiles = map_id_dict['tile_fetcher'].url_format,
    attr = 'Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    name = name,
    overlay = True,
    control = True
  ).add_to(self)

# Inicializando o Folium
folium.Map.add_ee_layer = add_ee_layer
folium.Map.add_ee_layer_vect = add_ee_layer_vect

# Dicionario com parametros de visualização
vis_params = {
  'min': 0,
  'max': 0.7,
  'bands':['B5','B4','B3']
  }

# Cria um objeto do tipo folium(Mapa) e inicia com localização e zoom defaults
meuMapa = folium.Map(location=[20, 0], zoom_start=3)

#Adiciona Camada ao Mapa
meuMapa.add_ee_layer(imagem1, vis_params, 'Imagem L5')

# Habilita opções interativas do mapa
meuMapa.add_child(folium.LayerControl())

# Exibe o Mapa
display(meuMapa)
```

```python
# EE.FEATURECOLLECTIOn
lsib = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017')
lsib_draw = lsib.draw('FF0000',5,2)
#obs.. um dos melhores jeitos de se visualizar um vetor é convertendo ele para imagem, veja o uso da função draw



#Map Clear
meuMapa = folium.Map(location=[20, 0], zoom_start=3)

meuMapa.add_ee_layer(lsib_draw, {'bands':['vis-red','vis-green','vis-blue'], min:0,max:255}, 'LSIB')
# Habilita opções interativas do mapa
meuMapa.add_child(folium.LayerControl())
# Exibe o Mapa
display(meuMapa)
```

```python
# EE.FEATURE
brasil = lsib.filterMetadata('country_co','equals','BR')
print(brasil.first().propertyNames().getInfo())
print(brasil.first().get('country_co').getInfo())

# Convertendo os varios poligonos do brasil em uma unica feature
brasil = brasil.union(1).first()


#Map Clear
meuMapa = folium.Map(location=[20, 0], zoom_start=3)

meuMapa.add_ee_layer_vect(brasil, {'color':'green'}, 'LSIB/Brasil') # Repare que aqui está usando a função add_ee_layer_vect, nao convertendo para imagem
# Habilita opções interativas do mapa
meuMapa.add_child(folium.LayerControl())
# Exibe o Mapa
display(meuMapa)
```

1.3.7 - Visualização de Graficos com Matplotlib


```python
# Importar a Biblioteca
import matplotlib.pyplot as plt

# Carregando uma ee.Image
img = ee.Image('LANDSAT/LT05/C01/T1_SR/LT05_034033_20000913')

# Seleção de Bandas e sorteio dos 500 pixeis aleatoriamente selcionados
samples = img.select(['B3','B4']).divide(10000).sample(scale=30, numPixels=500)
print(samples.limit(1).getInfo())


# Reduz os metadados B3 e B4 das amostras, para o tipo ee.List() com o uso do ee.Reducer.toList()
samp_dict = samples.reduceColumns(ee.Reducer.toList().repeat(2), ['B3', 'B4'])
print(samp_dict.getInfo())

samp_list = ee.List(samp_dict.get('list'))
print(samp_list.getInfo())

# Traz(Materializa) a lista do servidor para o cliente
samp_data = samp_list.getInfo()

# Exibe o scatter plot de RED em X e NIR em Y.
plt.scatter(samp_data[0], samp_data[1], alpha=0.2)
plt.xlabel('Red', fontsize=12)
plt.ylabel('NIR', fontsize=12)
plt.show()
```
